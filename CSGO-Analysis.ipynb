```python
import pandas as pd
```

```python
df = pd.read_csv('CSGOComplete.csv')
df.head()
```

```python
# unique map dan jumlah tiap unique
# hanya map dengan jumlah > 100 match yang akan dianalisis
# yaitu Dust II, Mirage dan Inferno
df.Map.value_counts()
```

```python
# map mirage
mirage = df.loc[df['Map'] == 'Mirage']
mirage
```

```python
# map dust II
dust_2 = df.loc[df['Map'] == 'Dust II']
dust_2
```

```python
# map inferno
inferno = df.loc[df['Map'] == 'Inferno']
inferno
```

```python
# presentase hasil tim A di map Mirage
win = len(mirage[mirage.Result == 'Win'])
tie = len(mirage[mirage.Result == 'Tie'])
lose = len(mirage[mirage.Result == 'Lost'])
total = win + tie + lose
print('Persen Kemenangan Tim A:', round(win * 100 / total, 2))
print('Persen Seri Tim A:', round(tie * 100 / total, 2))
print('Persen Kekalahan Tim A:', round(lose * 100 / total, 2))
```

```python
# presentase hasil tim A di map Dust II
win = len(dust_2[dust_2.Result == 'Win'])
tie = len(dust_2[dust_2.Result == 'Tie'])
lose = len(dust_2[dust_2.Result == 'Lost'])
total = win + tie + lose
print('Persen Kemenangan Tim A:', round(win * 100 / total, 2))
print('Persen Seri Tim A:', round(tie * 100 / total, 2))
print('Persen Kekalahan Tim A:', round(lose * 100 / total, 2))
```

```python
# presentase hasil tim A di map Inferno
win = len(inferno[inferno.Result == 'Win'])
tie = len(inferno[inferno.Result == 'Tie'])
lose = len(inferno[inferno.Result == 'Lost'])
total = win + tie + lose
print('Persen Kemenangan Tim A:', round(win * 100 / total, 2))
print('Persen Seri Tim A:', round(tie * 100 / total, 2))
print('Persen Kekalahan Tim A:', round(lose * 100 / total, 2))
```
